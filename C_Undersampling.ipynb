```python
# CSV 파일 읽기
normal_df = pd.read_csv('normal.csv')
hazard_df = pd.read_csv('hazard.csv')
```

```python
# 10초 단위 묶음 생성 함수
def create_10sec_groups_per_vehicle(df):
    # Group_ID를 행 단위로 생성
    df['Group_ID'] = (df.index // 10).astype(int)
    return df
```

```python
# Normal과 Hazard 데이터에 대해 10초 그룹 생성
normal_df = create_10sec_groups_per_vehicle(normal_df)
hazard_df = create_10sec_groups_per_vehicle(hazard_df)
```

# **언더샘플링**

```python
# 언더샘플링 함수
def balance_datasets_by_group(normal_df, hazard_df, ratio=2):
    """
    Normal 데이터에서 비율에 맞는 Group_ID를 샘플링하고 Hazard 데이터와 결합

    Parameters:
    - normal_df (pd.DataFrame): Normal 데이터프레임
    - hazard_df (pd.DataFrame): Hazard 데이터프레임
    - ratio (int): Normal:Hazard 비율 (기본값: 2)

    Returns:
    - pd.DataFrame: 비율에 맞게 결합된 데이터프레임
    """
    # 1. Hazard 데이터의 Group_ID 개수 확인
    hazard_group_count = hazard_df['Group_ID'].nunique()
    print(f"Hazard Group_ID 개수: {hazard_group_count}")

    # 2. Normal 데이터에서 필요한 Group_ID 개수 계산
    required_normal_groups = hazard_group_count * ratio
    print(f"Normal에서 샘플링할 Group_ID 개수: {required_normal_groups}")

    # 3. Normal 데이터에서 Group_ID 샘플링
    sampled_normal_group_ids = random.sample(
        normal_df['Group_ID'].unique().tolist(),
        k=required_normal_groups
    )

    # 4. Normal 데이터에서 선택된 Group_ID에 해당하는 데이터 가져오기
    sampled_normal_df = normal_df[normal_df['Group_ID'].isin(sampled_normal_group_ids)]
    print(f"샘플링된 Normal 데이터 개수: {len(sampled_normal_df)}")

    # 5. Normal 데이터와 Hazard 데이터 결합
    balanced_df = pd.concat([sampled_normal_df, hazard_df], ignore_index=True)

    # 6. Group_ID와 시간 순서로 정렬
    balanced_df = balanced_df.sort_values(['Group_ID', 'ISSUE_DATE']).reset_index(drop=True)

    print(f"최종 데이터 크기: {balanced_df.shape}")
    return balanced_df
```

```python
# 언더샘플링 데이터셋 생성
balanced_df = balance_datasets_by_group(normal_df, hazard_df, ratio=2)
```
